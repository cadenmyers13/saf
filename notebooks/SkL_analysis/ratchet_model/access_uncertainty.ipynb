```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npz file
for i, (dataset, numors) in enumerate(folder_movies.items()):
    for numor in numors:
        data = np.load(f"data/offsets_{numor}.npz")
       
        # Access the uncertainties
        uncertainty1 = data["uncertainty1"]
        uncertainty2 = data["uncertainty2"]
       
        # Plot histograms
        plt.figure(figsize=(12, 6))
       
        # Plot for uncertainty1
        plt.subplot(1, 2, 1)
        plt.hist(uncertainty1, bins=30, alpha=0.7, color='blue')
        plt.title('Histogram of Uncertainty 1')
        plt.xlabel('Uncertainty 1')
        plt.ylabel('Frequency')
       
        # Plot for uncertainty2
        plt.subplot(1, 2, 2)
        plt.hist(uncertainty2, bins=30, alpha=0.7, color='green')
        plt.title('Histogram of Uncertainty 2')
        plt.xlabel('Uncertainty 2')
        plt.ylabel('Frequency')
       
        # Show the plots
        plt.tight_layout()
        plt.show()
```
